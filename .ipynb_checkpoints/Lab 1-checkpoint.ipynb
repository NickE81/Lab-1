```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Task 1
df = pd.read_csv("flights.csv", na_values = "NaN")
# Clean data
df_clean = df[~df["arr_delay"].isnull()]

df_clean.head()
```

```python
# Task 2
# Find data for reports, graph next steps
# Airlines with most delayed departures on average
delayed_dep = df_clean[["carrier", "dep_delay"]].groupby(["carrier"]).mean().sort_values("dep_delay", ascending = False)
# Airlines with most delayed arrival times on average
delayed_arr = df_clean[["carrier", "arr_delay"]].groupby(["carrier"]).mean().sort_values("arr_delay", ascending = False)
# Months with most delayed arrivals on average
delayed_months = df_clean[["month", "arr_delay"]].groupby(["month"]).mean()
# Scheduled depature times with the most delay on average
delayed_times = df_clean[["sched_dep_time", "dep_delay"]].groupby(["sched_dep_time"]).mean()
# Average departure delay and arrival delay by origin of flight
delayed_origins = df_clean[["origin", "dep_delay", "arr_delay"]].groupby(["origin"]).mean()
```

```python
# Average departure delay by airline sorted in descending order
df_dep = pd.DataFrame(delayed_dep.index)
df_dep["dep_delay"] = delayed_dep.values
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
ax.table(cellText = df_dep.values, colLabels = df_dep.columns, loc = "center")
fig.tight_layout()
plt.show()
# Both Delta and United Airlines are better than average in terms of average departure delays, but Delta Airlines is lower.
```

```python
# Average arrival delay by airline sorted in descending order
df_arr = pd.DataFrame(delayed_arr.index)
df_arr["arr_delay"] = delayed_arr.values
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
ax.table(cellText = df_arr.values, colLabels = df_arr.columns, loc = "center")
fig.tight_layout()
plt.show()
# Both Delta and United Airlines are better than average in terms of average arrival delays, but Delta Airlines is lower.
```

```python
# Average arrival delay by month
plt.bar(delayed_months.index, height = delayed_months["arr_delay"])
plt.xlabel("Month")
plt.ylabel("Average Arrival Delay (mins)")
plt.show()
# June, July and December have the highest average delays in arrival, which could have to do with those months being common for vacations.
```

```python
# Average departure delay by time of day
plt.scatter(delayed_times.index, delayed_times["dep_delay"])
plt.xlabel("Scheduled Departure Time (hhmm)")
plt.ylabel("Average Departure Delay (mins)")
plt.show()
# The scatter plot appears to show that average departure delay tends to increase the later into the day that the flight is scheduled to depart.
```

```python
# Average departure and arrival delay by origin of the flight
df_origin = pd.DataFrame(delayed_origins.index)
df_origin["dep_delay"] = delayed_origins["dep_delay"].values
df_origin["arr_delay"] = delayed_origins["arr_delay"].values
fig, ax = plt.subplots()
fig.patch.set_visible(False)
ax.axis('off')
ax.axis('tight')
ax.table(cellText = df_origin.values, colLabels = df_origin.columns, loc = "center")
fig.tight_layout()
plt.show()
# Newark Liberty (EWR) has the highest average depature and arrival delays, and LaGuardia (LGA) and JFK have similar average arrival delays, but JFK has somewhat higher average departure delays.
```

```python
# Task 3
# Extract (max of) 20000 entries from data for United Airlines
dfUA = df_clean[df_clean["carrier"] == "UA"].head(20000)
# Extract (max of) 20000 entries from data for Delta Airlines
dfDL = df_clean[df_clean["carrier"] == "DL"].head(20000)
# Concatinate the two data sets
dfCom = pd.concat([dfUA, dfDL])
# Choose only carrier and arr_delay to be chosen as columns
pop_data = dfCom[["carrier", "arr_delay"]]

print(pop_data)
```

```python
# Task 4
# Get a sample of 1000 entries from Delta Airlines data
dl = pop_data[pop_data["carrier"] == "DL"].sample(n = 1000, random_state = 1)
# Give this data a sample_id of 1
dl["sample_id"] = 1
# Get a sample of 1000 entries from United Airlines data
ua = pop_data[pop_data["carrier"] == "UA"].sample(n = 1000, random_state = 1)
# Give this data a sample_id of 2
ua["sample_id"] = 2
```

```python
# Task 5
# Create a sample data set by concatinating the 2 samples taken in task 4
samples = pd.concat([dl, ua])
# Display the first 5 and last 5 elements of samples
samples.head(5), samples.tail(5)
# DL flights have sample_id 1 and UA flights have sample_id 2, as desired
```

```python
# Task 6
# Get the sample for Delta Airlines
samples_dl = samples[samples["sample_id"] == 1]
# Calculate mean for Delta Airlines sample
dl_mean = samples_dl[["arr_delay"]].mean()
# Calculate standard deviation for Delta Airlines sample
dl_std = samples_dl[["arr_delay"]].std()
# Calculate standard error for Delta Airlines sample
dl_se = dl_std / np.sqrt(1000)

# Get the sample for United Airlines
samples_ua = samples[samples["sample_id"] == 2]
# Calculate mean for United Airlines sample
ua_mean = samples_ua[["arr_delay"]].mean()
# Calculate standard deviation for United Airlines sample
ua_std = samples_ua[["arr_delay"]].std()
# Calculate standard error for United Airlines sample
ua_se = ua_std / np.sqrt(1000)
# Print the results with the desired format
print("United Airlines: mean " + str(ua_mean["arr_delay"]) + " and standard error " + str(ua_se["arr_delay"]))
print("Delta Airlines: mean " + str(dl_mean["arr_delay"]) + " and standard error " + str(dl_se["arr_delay"]))
```

```python
# Task 7
# Calculate lower value of CI for United Airlines sample
ua_lower = ua_mean["arr_delay"] - (1.96 * ua_se["arr_delay"])
# Calculate upper value of CI for United Airlines sample
ua_upper = ua_mean["arr_delay"] + (1.96 * ua_se["arr_delay"])
# Calculate lower value of CI for Delta Airlines sample
dl_lower = dl_mean["arr_delay"] - (1.96 * dl_se["arr_delay"])
# Calculate upper value of CI for Delta Airlines sample
dl_upper = dl_mean["arr_delay"] + (1.96 * dl_se["arr_delay"])
# Print confidences intervals
print(str(ua_lower) + " - " + str(ua_mean["arr_delay"]) + " - " + str(ua_upper) + "\n" +
      str(dl_lower) + " - " + str(dl_mean["arr_delay"]) + " - " + str(dl_upper))
```

```python
# Task 8
stats.ttest_ind(samples_dl[["arr_delay"]], samples_ua[["arr_delay"]])
# The p-value for our T-test is ~0.118 > 0.05, so we conclude that the mean arrival delay of the 2 data sets are not significantly different.
# We can not say whether or not Delta Airlines flights are more delayed than United Airlines flights.
```
